# Television

```python
from scipy import stats 
est = stats.linregress(x = df["TV"], y= df["Sales"])
est
```

```python
def predict(x):
    return est.slope * x + est.intercept
```

```python
r2_score(df["Sales"], df["TV"].apply(predict))
```

```python
mean_squared_error(df["Sales"], df["TV"].apply(predict))**0.5
```

```python
L.fit(df[["TV"]] , df["Sales"]) #L.fit(x,y)
```

```python
pred = L.predict(df[["TV"]])
pred
```

```python
r2_score(df["Sales"],pred)
```

```python
mean_squared_error(df["Sales"],pred)**0.5
```

```python
plt.plot(df[["TV"]], df["Sales"]) # plot first line
plt.plot(df[["TV"]], pred)
```

```python
fig, ax1 = plt.subplots(figsize=(12,  10))

# Instantiate a second axes that shares the same x-axis
ax2 = ax1.twinx()  
ax2.set_ylim(4, 30)
#ax1.plot(df[["TV"]], df["Sales"] , color="green")

ax1.set_xlabel("Investmnet of TV advertisemnet" ,)
plt.ylabel("Sales")

plt.scatter(df[["TV"]] , df["Sales"])
ax2.plot(df[["TV"]], pred , lw=4 , color = "red")
```
